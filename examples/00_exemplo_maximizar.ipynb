### Imports

```python
import gurobipy as gp
```

### Criando o modelo

```python
modelo = gp.Model('Exemplo_00')
```

### Inserindo as variáveis de decisão

```python
x1 = modelo.addVar()
x2 = modelo.addVar()
x3 = modelo.addVar()
```

### Definindo a função objetivo

```python
modelo.setObjective(12*x1 + 18*x2 + 22*x3, sense=gp.GRB.MAXIMIZE)
```

### Insere as restrições do modelo

```python
c1 = modelo.addConstr(1.5*x1 + 1.2*x3 <= 120)
c2 = modelo.addConstr(2.2*x2 + 1.4*x3 <= 200)
c3 = modelo.addConstr(1.2*x1 + 2*x2 +2.4*x3 <= 250)
```

### Resolvendo o modelo

```python
modelo.optimize()
```

### Acessando a solução

```python
print('x1: %i | x2: %i | x3: %i'%(x1.X, x2.X, x3.X))
```
